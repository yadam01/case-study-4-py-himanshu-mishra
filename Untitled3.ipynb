```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the Excel file
file_path = '/'
data = pd.read_excel(file_path)

# Step 2: Calculate percentages (Assuming marks are out of 15)
data['Percentage'] = (data['Marks'] / 15) * 100

# Step 3: Categorize students
greater_75 = data[data['Percentage'] > 75]
between_60_75 = data[(data['Percentage'] >= 60) & (data['Percentage'] <= 75)]
less_60 = data[data['Percentage'] < 60]

# Step 4: Print categorized data
print("Students with > 75% marks:\n", greater_75)
print("\nStudents with 60-75% marks:\n", between_60_75)
print("\nStudents with < 60% marks:\n", less_60)

# Step 5: Plot Histogram
plt.hist(data['Percentage'], bins=[0, 60, 75, 100], edgecolor='black', color='blue')
plt.xlabel('Percentage')
plt.ylabel('Number of Students')
plt.title('Histogram of Student Percentages')
plt.show()

# Step 6: Plot Scatter Plot
plt.scatter(data.index, data['Percentage'], color='red')
plt.xlabel('Students')
plt.ylabel('Percentage')
plt.title('Scatter Plot of Student Percentages')
plt.show()
```
